```python
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df
```

```python
#age distribution
sns.histplot(df,x='age',kde=True ,color='red')      #Adding kde(kerner density estiomate) for smooth curve
                                                    # representaion of singleplot
```

```python
#counts of pasenger class

sns.countplot(df,x='pclass')
```

```python
sns.countplot(df,x='pclass',hue='sex',palette='viridis') # pallete shows the colour
```

```python
#average fair by passenger class and gender
fig=sns.barplot(df,x='class',y='fare',hue='sex',palette='viridis')
fig.set_title("average fare by passenger class and gender")
fig.set_xlabel('passenger class')
fig.set_ylabel('average fare')
```
